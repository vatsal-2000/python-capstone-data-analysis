```python
import pandas as pd
import numpy as np

```

```python
# Task 1 : To make three dataframes in python and save them as three .csv files 

# Employee DataFrame
employee = pd.DataFrame({ 
    "ID": ["A001","A002","A003","A004","A005"],
    "Name": ["John Alter","Alice Luxumberg","Tom Sabestine","Nina Adgra","Amy Johny"],
    "Gender": ["M","F","M","F","F"],
    "City": ["Paris","London","Berlin","Newyork","Madrid"],
    "Age": [25,27,29,31,30] })

# Seniority DataFrame
seniority = pd.DataFrame({
    "ID": ["A001","A002","A003","A004","A005"],
    "Designation Level": [2,2,3,2,3] })

# Project DataFrame
project = pd.DataFrame({
    "ID": ["A001","A002","A003","A004","A005","A002","A005","A003","A001","A003","A001","A004","A004","A005"],
    "Project": ["Project 1","Project 2","Project 3","Project 4","Project 5","Project 6","Project 7","Project 8","Project 9","Project 10","Project 11","Project 12","Project 13","Project 14"],
    "Cost": [1002000,2000000,4500000,5500000,np.nan,680000,400000,350000,np.nan,300000,2000000,1000000,3000000,200000],
    "Status": ["Finished","Ongoing","Finished","Ongoing","Finished","Failed","Finished","Failed","Ongoing","Finished","Failed","Ongoing","Finished","Finished"] })
```

```python
# Saving dataframes as CSV files
employee.to_csv("employee.csv", index=False)
seniority.to_csv("seniority.csv", index=False)
project.to_csv("project.csv", index=False)
```

```python
# Reloading data from CSV files
employee = pd.read_csv("employee.csv")
seniority = pd.read_csv("seniority.csv")
project = pd.read_csv("project.csv")
```

```python
# Task 2 : To replace missing values for running average
running_sum = 0
count = 0

# Loop through each row
for i in range(len(project)):
    
# If value is NOT missing
    if not pd.isna(project.loc[i, "Cost"]): 
        running_sum += project.loc[i, "Cost"]
        count += 1
    
# If value is missing
    else:
        project.loc[i, "Cost"] = running_sum / count

project
```

```python
# Task 3 : To split the name into First and Last
employee[["First Name","Last Name"]] = employee["Name"].str.split(" ", expand=True)

# Drop original column
employee.drop("Name", axis=1, inplace=True)

employee
```

```python
# Task 4 : To merge all datasets in to one single dataframe
final = pd.merge(employee, seniority, on="ID")
final = pd.merge(final, project, on="ID")

final
```

```python
# Task 5 : To add bonus column
final["Bonus"] = np.where(final["Status"] == "Finished",
                          final["Cost"] * 0.05,
                          0)

final
```

```python
# Task 6 : Demote employees with failed projects

final.loc[final["Status"] == "Failed", "Designation Level"] += 1

# Remove employees with designation > 4
final = final[final["Designation Level"] <= 4]

final
```

```python
# Task 7 : To add Mr. & Mrs.
final["First Name"] = np.where(final["Gender"] == "M",
                              "Mr. " + final["First Name"],
                              "Mrs. " + final["First Name"])

# Drop Gender column
final.drop("Gender", axis=1, inplace=True)

final
```

```python
# Task 8 : To promote if age > 29

final.loc[final["Age"] > 29, "Designation Level"] -= 1
final
```

```python
# Task 9 : To group and calculate total cost

total_proj_cost = final.groupby(["ID","First Name"])["Cost"].sum().reset_index()

# Rename column
total_proj_cost.rename(columns={"Cost": "Total Cost"}, inplace=True)

total_proj_cost
```

```python
# Task 10 : Filtering cities with 'o'

result = final[final["City"].str.contains("o", case=False)]

result
```
